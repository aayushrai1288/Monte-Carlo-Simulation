```python
# Import required libraries
import numpy as np
import matplotlib.pyplot as plt
```

## Question 1: Expected Value and Variance in Particle Motion

A physicist measures the speeds of nitrogen gas particles at room temperature.

### Part (a): Calculate the average speed

```python
# Recorded speeds of nitrogen gas particles in m/s
speeds = np.array([420, 445, 380, 510, 398, 475, 425, 390, 460, 440])

# Calculate average speed using numpy mean function
avg_speed = np.mean(speeds)

# Print the result
print(f"Average speed = {avg_speed} m/s")
```

### Part (b): Calculate the standard deviation

```python
# Calculate standard deviation using numpy std function
std_speed = np.std(speeds)

# Print the result
print(f"Standard deviation = {std_speed} m/s")
```

### Part (c): Create a bar plot with average speed line

```python
# Create particle numbers from 1 to 10
particle_numbers = np.arange(1, 11)

# Create bar plot
plt.figure(figsize=(10, 6))
plt.bar(particle_numbers, speeds, color='skyblue', edgecolor='black')

# Add horizontal line at average speed
plt.axhline(y=avg_speed, color='red', linestyle='-', linewidth=2, label=f'Average = {avg_speed} m/s')

# Add labels and title
plt.xlabel('Particle Number')
plt.ylabel('Speed (m/s)')
plt.title('Particle Speed Distribution with Mean')
plt.legend()
plt.grid(axis='y', alpha=0.3)

# Display the plot
plt.show()
```

## Question 2: Normal Distribution and Probability Calculation

A temperature sensor takes 100 measurements following a normal distribution.

### Part (a): Generate random temperature measurements

```python
# Set random seed for reproducibility
np.random.seed(42)

# Generate 100 temperature measurements with mean=22 and std=0.5
temperatures = np.random.normal(loc=22, scale=0.5, size=100)

print(f"Generated {len(temperatures)} temperature measurements")
```

### Part (b): Create a histogram of temperature distribution

```python
# Create histogram with 20 bins
plt.figure(figsize=(10, 6))
plt.hist(temperatures, bins=20, color='lightgreen', edgecolor='black')

# Add labels and title
plt.xlabel('Temperature (°C)')
plt.ylabel('Frequency')
plt.title('Distribution of Temperature Measurements')
plt.grid(axis='y', alpha=0.3)

# Display the plot
plt.show()
```

### Part (c): Calculate percentage of measurements in range

```python
# Count measurements between 21.5°C and 22.5°C
count = np.sum((temperatures >= 21.5) & (temperatures <= 22.5))

# Calculate percentage
percentage = (count / 100) * 100

# Print the result
print(f"Percentage of measurements between 21.5°C and 22.5°C = {percentage}%")
```

## Question 3: Law of Large Numbers

Demonstrating that averaging many particles gives a stable result.

### Part (a): Generate random velocity values

```python
# Set random seed for reproducibility
np.random.seed(42)

# Generate 5000 random velocity values with scale=2.5
# Using Rayleigh distribution which is common for velocity magnitudes
velocities = np.random.rayleigh(scale=2.5, size=5000)

print(f"Generated {len(velocities)} velocity values")
```

### Part (b): Calculate averages for different sample sizes

```python
# Calculate average for first 100, 1000, 3000, and 5000 velocities
avg_100 = np.mean(velocities[:100])
avg_1000 = np.mean(velocities[:1000])
avg_3000 = np.mean(velocities[:3000])
avg_5000 = np.mean(velocities[:5000])

# Store in a list
averages = [avg_100, avg_1000, avg_3000, avg_5000]

print(f"Averages: {averages}")
```

### Part (c): Create line plot showing convergence

```python
# List of sample sizes
sample_sizes = [100, 1000, 3000, 5000]

# Create line plot
plt.figure(figsize=(10, 6))
plt.plot(sample_sizes, averages, marker='o', linewidth=2, markersize=8, color='blue')

# Add horizontal dashed red line at final average
plt.axhline(y=avg_5000, linestyle='--', color='red', linewidth=2, label=f'Final Average = {avg_5000:.3f}')

# Add labels and title
plt.xlabel('Number of Samples')
plt.ylabel('Average Velocity')
plt.title('Law of Large Numbers: Convergence of Average')
plt.legend()
plt.grid(True, alpha=0.3)

# Display the plot
plt.show()
```

## Question 4: Comparing Two Distributions

Comparing measurements from two different instruments.

### Part (a): Generate measurements from both instruments

```python
# Set random seed for reproducibility
np.random.seed(42)

# Generate 500 measurements from Instrument 1: mean=5.80, std=0.15
inst1 = np.random.normal(loc=5.80, scale=0.15, size=500)

# Generate 500 measurements from Instrument 2: mean=5.75, std=0.35
inst2 = np.random.normal(loc=5.75, scale=0.35, size=500)

print(f"Generated {len(inst1)} measurements for Instrument 1")
print(f"Generated {len(inst2)} measurements for Instrument 2")
```

### Part (b): Create overlapping histograms

```python
# Create overlapping histograms with transparency
plt.figure(figsize=(10, 6))
plt.hist(inst1, bins=15, color='blue', alpha=0.6, label='Instrument 1', edgecolor='black')
plt.hist(inst2, bins=15, color='red', alpha=0.6, label='Instrument 2', edgecolor='black')

# Add labels and title
plt.xlabel('Conductivity (S/m)')
plt.ylabel('Frequency')
plt.title('Comparison of Two Instruments')
plt.legend()
plt.grid(axis='y', alpha=0.3)

# Display the plot
plt.show()
```

### Part (c): Calculate precision ratio

```python
# Calculate standard deviations
std_inst1 = np.std(inst1)
std_inst2 = np.std(inst2)

# Calculate ratio of standard deviations
precision_ratio = std_inst2 / std_inst1

# Print the result with ratio rounded to 2 decimal places
print(f"Instrument 2 is {precision_ratio:.2f} times more scattered than Instrument 1")
```
